```python
import numpy as np
import matplotlib.pyplot as plt 

N=10
x=np.linspace(0,10,N)# 0 dan 10 a kadar 10 tane eşit aralıkta sayı
y=x**2
plt.plot(x,y,"o--")
# plt.axis("off")#eksenleri kapatır
plt.grid(True)#ızgara oluşturuyor
plt.axis([0,50,0,100])#aralık belirle
plt.legend(["x**2"]) #sol tarafta renkleri anlatır
plt.xlabel("x kenarı")#x eksenine yazar
plt.ylabel("y kenarı")#y eksenine yazar
plt.show()


```

```python
import cv2 
import numpy as np 


def findMaxContours(contours):
    max_i=0
    max_area=0
    for i in range(len(contours)):
        area_face=cv2.contourArea(contours[i])
        if area_face>max_area:
            max_area=area_face
            max_i=i
    try:
        cnt=contours[max_i]
    except:
        contours=[0]
        cnt=contours[0]
    return cnt


cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    frame=cv2.flip(frame,1)
    roi=frame[50:300,200:450]#frame[y1:y2,x1:x2]
    cv2.rectangle(frame,(200,50),(450,300),(0,0,255),0)

    hsv=cv2.cvtColor(roi,cv2.COLOR_BGR2HSV)
    lower_color=np.array([0,45,79],np.uint8)
    upper_color=np.array([17,255,255],np.uint8)
    mask=cv2.inRange(hsv,lower_color,upper_color)
    kernel=np.ones((3,3),np.uint8)
    mask=cv2.dilate(mask,kernel,iterations=1)
    mask=cv2.medianBlur(mask,15)


    contours,_=cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)


    if len(contours)>0:
        try:
            cnt=findMaxContours(contours)
            exLeft=tuple(cnt[cnt[:,:,0].argmin()][0])
            exRight =tuple(cnt[cnt[:,:,0].argmax()][0])
            exTop =tuple(cnt[cnt[:,:,1].argmin()][0])
            exBottom =tuple(cnt[cnt[:,:,1].argmax()][0])
            cv2.circle(roi,exLeft,5,(0,255,0),2)
            cv2.circle(roi,exRight,5,(0,255,0),2)
            cv2.circle(roi,exTop,5,(0,255,0),2)
            cv2.circle(roi,exBottom,5,(0,255,0),2)
            cv2.line(roi,exTop,exLeft,1,2)
            cv2.line(roi,exTop,exRight,1,2)
            cv2.line(roi,exBottom,exLeft,1,2)
            cv2.line(roi,exBottom,exRight,1,2)







        except:
            pass







    cv2.imshow("1",frame)
    cv2.imshow("2",roi)
    cv2.imshow("3",mask)
    
    

    if cv2.waitKey(20) & 0xFF==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np 
# işaretçimizin Renginin aralığını belirliyoruz
lower_blue=(60,74,36)
upper_blue=(113,255,255)
cap=cv2.VideoCapture(0)

paintWindow = np.zeros((480,640,3))+255
#renklerimizin ekranda kapladığı pixeller
black=[]
yellow=[]
red=[]
colors=[(0,0,0),(0,255,255),(0,0,255)]

colorIndex=None
while True:
    ret,frame=cap.read()
    if ret is not True:
        break
    frame=cv2.flip(frame,1)
    result=frame.copy()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    # her butonun alanını roi ile işaretliyoruz
    clear_roi=frame[0:50,0:100]
    cv2.rectangle(clear_roi,(0,0),(100,50),(255,255,255),-1)
    
    black_roi=frame[0:50,100:200]
    cv2.rectangle(black_roi,(0,0),(100,50),(0,0,0),-1)

    yellow_roi=frame[0:50,200:300]
    cv2.rectangle(yellow_roi,(0,0),(100,50),(0,255,255),-1)

    red_roi=frame[0:50,300:400]
    cv2.rectangle(red_roi,(0,0),(100,50),(0,0,255),-1)


    # işaretci konturunu bulyoruz
    mask=cv2.inRange(hsv,lower_blue,upper_blue)
    mask = cv2.erode(mask,(5,5),iterations =2)

    contours,hierarchy=cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    




# kontur var mı diye bakıyoruz
    if len(contours)>0 :
        # en büyük konturu alıyoruz
        cnt=max(contours,key=lambda x:cv2.contourArea(x))
        (x,y,w,h)=cv2.boundingRect(cnt)
        cursor=[x+w/2,y+h/2]        # işaretci kordinatları

        cv2.circle(frame,(int(cursor[0]),int(cursor[1])),10,(0,0,0),-1)#işaretciyi siyah daire ile göster
        # beyaz alandaysa silcek herşeyi o yüzden resim verisi olan arrayleri sıfırlıyoz
        if 100>cursor[0]>0 and 50>cursor[1]>0:
            black=[]
            yellow=[]
            red=[]
# renklere göre color index alıyorlar
        elif 200>cursor[0]>100 and 50>cursor[1]>0:
            colorIndex=0

        elif 300>cursor[0]>200 and 50>cursor[1]>0:
            colorIndex=1
        elif 400>cursor[0]>300 and 50>cursor[1]>0:
            colorIndex=2


# color indexlerine göre cursor datası pixel boyuyor
        if colorIndex==0:
            black.append(cursor)
        if colorIndex==1:
            yellow.append(cursor)
        if colorIndex==2:
            red.append(cursor)



    points=[black,yellow,red]
    


# boyama kısmı
    for i in range(len(points)):
        for j in range(len(points[i])):
            
            try:
               cv2.line(frame,(int(points[i][j][0]),int(points[i][j][1])),(int(points[i][j+1][0]),int(points[i][j+1][1])),colors[i],3)
            except:
                pass















    cv2.imshow("1",frame)
    cv2.imshow("2",mask)
    cv2.imshow("3",paintWindow)
    if cv2.waitKey(20) & 0xFF==ord("q"):
        break







print(points)








cap.release()
cv2.destroyAllWindows()
```
